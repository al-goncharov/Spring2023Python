## Exercise 1  
1. Создайте график $x^2⋅sin(1/x^2)+x$ на интервале $[−1,1]$, используя 250 точек. Не забудьте пометить оси!
2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы  функций $1/(1+x^2)$ и $1/x^2$ на интервале $[5,25]$. (Относительная разница чисел $a$ и $b$ определяется как $|1−a/b|$)
3. В Scipy также есть много менее распространенных специальных функций. Взгляните на обширный список <a href=https://docs.scipy.org/doc/scipy/reference/special.html> здесь</a>! Выберите несколько интересных спец функций, которые вы знаете (или, может быть, не знаете), подставьте несколько чисел и начертите их.


```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import scipy

xx = np.linspace(-1, 1, 250)

yy = (xx ** 2) * np.sin(1/(xx ** 2)) + xx
plt.plot(xx, yy)
```

```python
xx = np.linspace(5, 25, 250)
yy = abs(1 - (1/(1 + xx ** 2)) / (1 / (xx ** 2)))

plt.semilogy(xx, yy)
```

```python
xx = np.linspace(-5, 5, 50)
c = scipy.special.cbrt(xx)
w = scipy.special.lambertw(xx)
plt.plot(xx, c, label = 'Кубическая корневая функция')
plt.plot(xx, w, label = 'Функция Ламберта', color = 'red', linestyle = '--')
plt.legend()
```

```python
xx = np.linspace(-6*np.pi, 6*np.pi, 250)
yy = np.sin(xx) / xx
plt.figure(figsize=(15, 5))
x = np.arange(-6*np.pi, 6*np.pi + np.pi/2, np.pi/2)
y = []
for i in range(-12, 13):
    if i % 2 == 0:
        if i == 0:
            y.append('0')
        elif i == -2:
            y.append(r'$-\pi$')
        elif i == 2:
            y.append(r'$\pi$')
        else:
            y.append(r'$%s\pi$'%(i // 2))
    else:
        if i != 1 and i != -1:
            y.append(r'$\dfrac{%s\pi}{2}$'%(i))
        elif i == -1:
            y.append(r'$-\dfrac{\pi}{2}$')
        else:
            y.append(r'$\dfrac{\pi}{2}$')

plt.plot(xx, yy, label = 'sin(x)/x')
plt.grid()
plt.xticks(x, y)
plt.legend()
```

## Exercise 3  
1. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.
2. Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.

```python
x = np.linspace(-np.pi,np.pi,100)
y = np.linspace(-np.pi,np.pi,100)
X,Y = np.meshgrid(x,y)

f = np.sin(X) * np.sin(Y)
fig, ax = plt.subplots(figsize=(6,5))

p = ax.pcolor(X, Y, f, cmap='hot')

cb = fig.colorbar(p, ax=ax)
cb.set_ticks([-1, -.5, 0, .5, 1])
```

```python
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)
X,Y = np.meshgrid(x,y)

f = np.exp(-X**2 - Y**2)
fig, ax = plt.subplots(figsize=(6,5))

p = ax.pcolor(X, Y, f, norm =  mpl.colors.Normalize(-abs(f).max(), abs(f).max()), cmap='hot')

cb = fig.colorbar(p, ax=ax)
cb.set_ticks([-1, -.5, 0, .5, 1])
```
